## Features

**Ascites**: (מיימת) Is an objextive medical sign for water concentration in the Peritoneum (צפק). 84% of them are caused by Cirrhosis (שחמת הככבד). And the rest is due to benign tummors or Pancrea desiases.



**Triglyceride**: is an organic compound that occupies most of the fat stores in the body.
Their breakdown happens mainly during hunger.
Therefore nowadays a high concentration of them comes mainly with obesity,
And can cause pancreatic diseases and accumulation of fat in unusual places.
Normal values are: Less than 150 mg/dL

**Platelets**: are the smallest anucleated cell in the bloodstream whose main role is in the blood clotting process.
Normal platelet levels are: 150,000-450,000 platelets/microliter.
An abnormally high amount can result from inflammation, infections and pre-cancerous diseases.

**Prothrombin or Prothrombin Time** (PT): is the time it takes for the blood to clot.
Normal time is between 10-15 seconds. This test is done in order to, among other things, check the function of the liver, since the aforementioned protein is produced in the liver.

**Hepatomegaly**: An enlarged liver is one that's bigger than normal. The medical term is hepatomegaly (hep-uh-toe-MEG-uh-le). Rather than a disease, an enlarged liver is a sign of an underlying problem, such as liver disease, congestive heart failure or cancer,
values: Y/N meaning (Enlarged/Not Enlarged).

**Albumin**: Albumin is the most abundant protein in the blood. It is created in the liver, and its main function is to maintain the oncotic pressure - the pressure that keeps the fluids inside the blood vessels. In addition to this, the albumin helps transport various chemical substances - such as hormones, calcium and drugs - in the bloodstream.
Normal Values:  3.4 to 5.4 g/dL (34 to 54 g/L).

**Copper**: Although measuring the copper level in the blood, urine, and sometimes even in a liver biopsy is naturally identified with the confirmation or denial of Wilson's disease, measuring the copper level is also required to diagnose the rare Menkes disease, conditions of anemia or leukopenia or even neutropenia due to a chronic lack of copper. Even in some bone diseases and persistent osteoporosis, there is sometimes a requirement to measure the copper level in the blood.
Normal Values: 62 to 140 micrograms per deciliter (mcg/dL)

**Alk_Phos - Alkaline Phosphatase**:
The metabolic role of this enzyme is not entirely clear. It is related to the transfer of components through membranes, and also has a role in the calcification of the bone. The enzyme is found in various tissues in the body, mainly in the liver, bile ducts, intestine, bones and placenta.
The enzyme is released into the blood as a result of disease or when cells that contain it are damaged. The enzyme test is especially important for diagnosing liver and bone diseases.
The enzyme is also related to bone building, so during the growth process you see an increase in it in the blood.
Normal Values: 30 to 120 (IU/L), 501 to 2004 (µg/L)

**Age**

The number of the age is high, we can infer that the age is in days. -> We convert the age to years by divde the days in 365.


**Spiders**

Spider angiomas are a collection of blood vessels under the surface of your skin that resembles a spider. Angiomas are red to purple marks on your skin. Treatment isn’t necessary but your healthcare provider can remove the mark if you don’t like how it looks. If you have multiple angiomas, it could be a sign of an underlying condition.


**Edema**

Edema is the medical term for swelling caused by fluid trapped in your body's tissues. Edema happens most often in your feet, ankles and legs, but can affect other parts of your body, such as your face, hands and abdomen



**Bilirubin**

Bilirubin (bil-ih-ROO-bin) is a yellowish pigment that is made during the breakdown of red blood cells. Bilirubin passes through the liver and is eventually excreted out of the body.

Higher than usual levels of bilirubin may indicate different types of liver or bile duct problems. Sometimes, higher bilirubin levels may be caused by an increased rate of destruction of red blood cells.
Normal values: 0.3 to 1.2 milligrams per deciliter.





**Cholesterol**

Cholesterol is the principal sterol of all higher animals, distributed in body tissues, especially the brain and spinal cord, and in animal fats and oils.

Normal range:

Total Cholesterol < 200

LDL Cholesterol < 100

HDL Cholesterol >= 60

**SGOT - AST (GOT)**:
the concentration of the GOT enzyme in the blood. This enzyme is found mainly in the liver, muscles (including heart muscle) and red blood cells. The enzyme is released into the blood when cells that contain it are damaged.
Normal Values: 0 to 35 (IU/L)

```python
norm_min = 0.3
norm_max = 1.2
new_column_values = np.where((df['Bilirubin'] < norm_min), 'L',
                             np.where((df['Bilirubin'] > norm_max), 'H', 'IR'))
df.insert(df.columns.get_loc('Bilirubin') + 1, 'N_Bilirubin', new_column_values)
```

```python
norm_min = 3.4
norm_max = 5.4
new_column_values = np.where((df['Albumin'] < norm_min), 'L',
                             np.where((df['Albumin'] > norm_max), 'H', 'IR'))
df.insert(df.columns.get_loc('Albumin') + 1, 'N_Albumin', new_column_values)
```

```python
norm_min = 62
norm_max = 140
new_column_values = np.where((df['Copper'] < norm_min), 'L',
                             np.where((df['Copper'] > norm_max), 'H', 'IR'))
df.insert(df.columns.get_loc('Copper') + 1, 'N_Copper', new_column_values)
```

```python
norm_min = 501
norm_max = 2004
new_column_values = np.where((df['Alk_Phos'] < norm_min), 'L',
                             np.where((df['Alk_Phos'] > norm_max), 'H', 'IR'))
df.insert(df.columns.get_loc('Alk_Phos') + 1, 'N_Alk_Phos', new_column_values)
```

## Data Preparation

```python
final_columns = ['id',
 'N_Days',
 'Drug',
 'Age',
 'Sex',
 'Ascites',
 'Hepatomegaly',
 'Spiders',
 'Edema',
 'N_Bilirubin',
 'N_Cholesterol',
 'N_Albumin',
 'N_Copper',
 'N_Alk_Phos',
 'N_SGOT',
 'NTryglicerides',
 'NPlatelets',
 'NProthrombin',
 'Stage',
 'Status']
```

```python
target_column = "Status"
final_columns = set(final_columns)
non_target_columns = final_columns.difference([target_column])

df["Stage"] = df["Stage"].astype(str)
df_encoded = pd.get_dummies(df[non_target_columns])
df_m = pd.concat([df[target_column], df_encoded], axis=1)
df_m.head()
df_m.info()
```

```python
df_m.columns
```

```python
df_m.head()
```

```python
from sklearn.model_selection import train_test_split
y = df_m['Status']
X = df_m.drop(columns = ['Status'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## Machine Learning - Classification Models

### Naive Base Algorithm

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

naive = MultinomialNB()
naive.fit(X_train, y_train)
y_predicted = naive.predict(X_test)
accuracy = accuracy_score(y_true=y_test, y_pred=y_predicted)
print("Accuracy:", accuracy)
```

```python
results_df = pd.DataFrame({
    "predicted": y_predicted,
    "actual": y_test
})
results_df
```

###Gaussian process regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
```

```python
temp_df = df_m.copy()

# Check the unique values in the 'Status' column before mapping
print("Unique values in 'Status' column before mapping:", temp_df['Status'].unique())

# Define the status mapping dictionary
status_mapping = {'C': 0, 'D': 1, 'CL': 2}

# Map the values in the 'Status' column
temp_df['Status'] = temp_df['Status'].map(status_mapping)

# Check the unique values in the 'Status' column after mapping
print("Unique values in 'Status' column after mapping:", temp_df['Status'].unique())
```

```python
y = temp_df['Status']
X = temp_df.drop(columns = ['Status'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
kernel = 1.0 * RBF(length_scale=1.0)
model = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)
model.fit(X_train, y_train)
y_pred, sigma = model.predict(X_test, return_std=True)
```

### KNN Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier()
k_values = np.arange(1, 30)
param_grid = {'n_neighbors': k_values}
clf = GridSearchCV(knn, param_grid, cv=10, scoring='accuracy')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
results_df = pd.DataFrame({
    "predicted": y_pred,
    "actual": y_test
})

results_df
```
